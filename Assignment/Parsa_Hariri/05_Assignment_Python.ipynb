```python
from os import execlp
import urllib.request

def download_file(url, path):
  try:     
    if url[-4:] == ".txt":
      p = path + '.txt'
      urllib.request.urlretrieve(url, p)
    else: 
      raise NameError('URL name wrong name') 
  except:
    return "No text file found at given URL, download aborted!"
```

```python
import string
def normalize_text(text):
  text = text.lower()
  text = text.translate(str.maketrans('', '', string.punctuation))
  return text
```

```python
from logging import ERROR
import copy
def IsStopwords(word):
  list_of_stop_words = []
  with open('eng_stop_words.txt') as f:
    Lines = f.readlines()
    for line in Lines:
      list_of_stop_words.append(line.strip())
  if word in list_of_stop_words:
    return True
  else:
    return False

def remove_stopwords(word_list):
  #word_list is a list of words 
  try:
    wlist = copy.deepcopy(word_list)

    for word in word_list:
      if IsStopwords(word) or word == '':
        wlist.remove(word)
    return wlist
  except:
    return ERROR
```

```python
def tokenize_text(text):
  word_list = text.split(' ')
  while '' in word_list:
    word_list.remove('')
  return word_list
print(tokenize_text('parsa apspxap\ndcpcpd ccdcdc ccc     ccc c'))
```
